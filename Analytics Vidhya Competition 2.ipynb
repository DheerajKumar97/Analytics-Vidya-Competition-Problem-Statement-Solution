# Data Preprocessing

```python
df1['age'] = df1['age'].fillna(df1['age'].median())
df1['trainee_engagement_rating'] = df1['trainee_engagement_rating'].fillna(df1['trainee_engagement_rating'].median())
```

```python
df1['age'] =df1['age'].astype(int)
df1['trainee_engagement_rating'] = df1['trainee_engagement_rating'].astype(int)
```

```python
df1.dtypes
```

```python
df1 =df1.drop(['id'],axis=1)
```

```python
cat_df1 = df1.select_dtypes(include="object")
cat_df1
```

```python
Non_label_cat_df = cat_df1.loc[:,['difficulty_level','education']]
```

```python
Non_label_cat_df
```

```python
def encode_education(education):
    if education == "High School Diploma":
        return 5
    elif education == "Matriculation":
        return 4
    elif education == "Bachelors":
        return 3
    elif education == "No Qualification":
        return 2
    elif education == "Masters":
        return 1
    else:
        return 0
Non_label_cat_df['education'] = Non_label_cat_df['education'].apply(encode_education)
```

```python
def encode_difficulty_level(difficulty_level):
    if difficulty_level == "easy":
        return 4
    elif difficulty_level == "intermediate":
        return 3
    elif difficulty_level == "hard":
        return 2
    elif difficulty_level == "vary hard":
        return 1
    else:
        return 0
Non_label_cat_df['difficulty_level'] = Non_label_cat_df['difficulty_level'].apply(encode_difficulty_level)
```

```python
Non_label_cat_df
```

```python
cat_df1 = cat_df1.drop(['difficulty_level','education'],axis=1)
```

```python
cat_df1
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
cat_df1 = cat_df1.apply(lambda col: le.fit_transform(col))
cat_df1
```

```python
cat_df1 = pd.concat([cat_df1,Non_label_cat_df],axis=1)
cat_df1
```
