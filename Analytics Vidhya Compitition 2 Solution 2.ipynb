```python
def encode_education(education):
    if education == "High School Diploma":
        return 3
    elif education == "Matriculation":
        return 2
    elif education == "Bachelors":
        return 4
    elif education == "No Qualification":
        return 1
    elif education == "Masters":
        return 5
    else:
        return 0
cat_var['education'] = cat_var['education'].apply(encode_education)
```

```python
def encode_difficulty_level(difficulty_level):
    if difficulty_level == "easy":
        return 1
    elif difficulty_level == "intermediate":
        return 2
    elif difficulty_level == "hard":
        return 3
    elif difficulty_level == "vary hard":
        return 4
    else:
        return 0
cat_var['difficulty_level'] = cat_var['difficulty_level'].apply(encode_difficulty_level)
```

```python
cat_var['isPass'] =df1['is_pass']
cat_var
```

```python
mean_encode = cat_var.groupby('id')['isPass'].mean()
```

```python
cat_var.loc[:,'encoded_id'] = cat_var.id.map(mean_encode)
```

```python
mean = cat_var['isPass'].mean()
agg =  cat_var.groupby('id')['isPass'].agg(['count','mean'])
counts = agg['count']
means =agg['mean']
weight =100
```

```python
smooth = (counts * means + weight * means) / (counts + weight)
smooth
```

```python
cat_var.loc[:,'encoded_id'] = cat_var.id.map(smooth)
```

```python
cat_var = cat_var.drop(['id'],axis=1)
cat_var = cat_var.drop(['is_handicapped'],axis=1)
```

```python
cat_var
```

```python
cat_var['gender'] = pd.get_dummies(cat_var['gender'])
cat_var['test_type'] = pd.get_dummies(cat_var['test_type'])
```

```python
mean_encode = cat_var.groupby('program_id')['isPass'].mean()
cat_var.loc[:,'encoded_program_id'] = cat_var.program_id.map(mean_encode)
mean = cat_var['isPass'].mean()
agg =  cat_var.groupby('program_id')['isPass'].agg(['count','mean'])
counts = agg['count']
means =agg['mean']
weight =100
smooth = (counts * means + weight * means) / (counts + weight)
smooth
```

```python
cat_var.loc[:,'encoded_program_id'] = cat_var.program_id.map(smooth)
```

# Model Building

```python
x = df.drop([7],axis=1)
y = df[7]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression 
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix 
```

```python
pipeline_lr1=Pipeline([('lr_classifier',LogisticRegression(random_state=0))])
pipeline_dt1=Pipeline([('dt_classifier',DecisionTreeClassifier())])
pipeline_rf1=Pipeline([('rf_classifier',RandomForestClassifier())])
pipeline_Nb1=Pipeline([('Nb_Gaussion',GaussianNB())])
pipeline_XGb1=Pipeline([('XGb_classifier',XGBClassifier())])
```

```python
pipelines1 = [pipeline_lr1, pipeline_dt1, pipeline_rf1,pipeline_Nb1,pipeline_XGb1]
```

```python
best_accuracy1=0.0
best_classifier1=0
best_pipeline1=""
pipe_dict1 = {0: 'Logistic Regression', 1: 'Decision Tree', 2: 'RandomForest',3:'GaussianNB',4: 'XGboost'}
```

```python
for pipe in pipelines1:
    pipe.fit(x_train, y_train)
```

```python
for i,model in enumerate(pipelines1):
    print("{} Test Accuracy: {}".format(pipe_dict1[i],model.score(x_test,y_test)))
    print(classification_report(y_test,model.predict(x_test)))
```
