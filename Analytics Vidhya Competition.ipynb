## Model Building

```python
x = df.drop(['Sales'],axis=1)
y = df['Sales']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=40)
```

```python
import statsmodels.api as sm
model2 =sm.OLS(y_train,x_train).fit()
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression(copy_X=True, fit_intercept=True, n_jobs=0, normalize=False)
```

```python
LR =lr.fit(x_train,y_train)
```

```python
LR_Pred = lr.predict(x_test)
```

```python
from sklearn import metrics
LR_RMSE = np.sqrt(metrics.mean_squared_error(y_test,LR_Pred))
LR_RMSE
```

```python
from sklearn import metrics
LR_MSE = (metrics.mean_squared_error(y_test,LR_Pred))
LR_MSE
```

```python
from sklearn import metrics
LR_MAE=(metrics.mean_absolute_error(y_test, LR_Pred))
LR_MAE
```

```python
from sklearn.metrics import r2_score
LR_r2_score = r2_score(y_test,LR_Pred)
LR_r2_score
```
